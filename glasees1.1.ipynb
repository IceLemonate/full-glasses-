```python
#s初始化
a=[0,0]
b=[1,1]
c=[]
s=[]
s.append(a)
s.append(b)
s.append(c)
# print(s)
#获得杯子顶部颜色的长度,参数应该是一个杯子的深拷贝,调用这个函数的时候参数一定要是深拷贝！！！
def topLen(sb):
    ss=sb.copy()
    ta=ss[len(ss)-1]
    i=0
    while(len(ss)!=0 and ss[len(ss)-1]==ta):
        i+=1
        ss.pop()
    return i

#玩家开始玩的时候调用的函数
def full(a,b):
    la=len(s[a])#杯子中液体的长度
    lb=len(s[b])
    if(la==0):
        print("a is empty")
        return 0
    if(lb==4):
        print("already full")
        return 0
    ta=s[a][la-1]#杯子顶部液体颜色（数字）ta tb
    #获得杯子a顶层颜色长度tla
    acopy=s[a].copy()
    tla=topLen(acopy)
#     tla=topLen(a)
    if(lb!=0):
        tb=s[b][lb-1]
        if(ta!=tb):
            print("颜色不同无法倒入")
            return 0
    #不用再写循环，此时不然b为空瓶，不然表层颜色相同
    tlb=4-lb#b表层剩余的空间 tlb
    ma=min(tla,tlb)
    print("ma=",ma)
    print("ss",s)
    for i in range(0,ma):
        print("i=",i)
        topa=s[a].pop()
        s[b].append(topa)
        print(s)
    return 1

#对杯子进行打乱的时候调用的函数
def initFull(a,b):
    la=len(s[a])
    lb=len(s[b])
    if(la==0 or lb==4):
        print("a是空的或者b是满的")
        return 0
    t=s[a].pop()
    s[b].append(t)
    if(len(s[a])==0):
        global flag
        flag+=1  #如果有个杯子倒完了，则标记flag=1,结束init
        return a #满了则返回空地的杯子号
    return 9#成功

#检测是否已经初始化成功/是否全部是纯色，即成功过关
#先检测空杯子的数量，然后再检测非空杯子的纯度
def tstinit(ss):
    for i in range(0,len(ss)):
        if(len(ss[i])!=0 and topLen(ss[i])!=4):
           return 0#不是纯色返回失败
    return 1
```
